```python
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
from langchain_community.embeddings import HuggingFaceEmbeddings
from fastembed import TextEmbedding
embed_model = HuggingFaceEmbeddings(model_name = 'sentence-transformers/all-mpnet-base-v2')
```

```python
from google.colab import userdata
from groq import Groq
from langchain_groq import ChatGroq
groq_api_key = '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'
openai_api_key = 'sk-proj-yPOx6hbrSvCh4Lr0TCUrT3BlbkFJ7497z14zyXAt4loRqjyz'
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings

semantic_chunker = SemanticChunker(embed_model, breakpoint_threshold_type="percentile")
#
semantic_chunks = semantic_chunker.create_documents([d.page_content for d in documents])
print(semantic_chunks)
```

```python
from langchain_community.vectorstores import Chroma
semantic_chunk_vectorstore = Chroma.from_documents(semantic_chunks, embedding=embed_model)
```

```python
semantic_chunk_retriever = semantic_chunk_vectorstore.as_retriever()
```

```python
from langchain_core.prompts import ChatPromptTemplate

rag_template = """\
Use the following context to answer the user's query. If you cannot answer, please respond with 'I don't know'.

User's Query:
{question}

Context:
{context}
"""

rag_prompt = ChatPromptTemplate.from_template(rag_template)
```

```python
import openai
openai.api_key = openai_api_key
```

```python
chat_model = ChatGroq(temperature=0,
                      model_name="mixtral-8x7b-32768",
                      api_key=groq_api_key)
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

semantic_rag_chain = (
    {"context" : semantic_chunk_retriever, "question" : RunnablePassthrough()}
    | rag_prompt
    | chat_model
    | StrOutputParser()
)
```

```python
semantic_rag_chain.invoke("Describe the Feature-based Approach with BERT?")
```

```python
naive_chunk_vectorstore = Chroma.from_documents(naive_chunks, embedding=embed_model)
naive_chunk_retriever = naive_chunk_vectorstore.as_retriever(search_kwargs={"k" : 5})
naive_rag_chain = (
    {"context" : naive_chunk_retriever, "question" : RunnablePassthrough()}
    | rag_prompt
    | chat_model
    | StrOutputParser()
)
```

```python
synthetic_data_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=0,
    length_function=len,
    is_separator_regex=False
)
#
synthetic_data_chunks = synthetic_data_splitter.create_documents([d.page_content for d in documents])
```

```python
ground_truths_semantic = []
contexts = []
questions = ['In terms of numbers of cases, how did cases of mumps change from 1968 to 2011?',
             'Which muscle is the fibularis tertius associated with?',
             'According to the CDC, what percentage of injury-related deaths are caused by head injuries?',
             'Where does the sciatic nerve branch?',
             'How great is the force applied to each foot when running?']
answers = ['According to the U.S. Centers for Disease Control and Prevention (CDC), the number of mumps cases dropped from more than 150,000 in 1968 to 11 reported cases in 2011',
           'The fibularis tertius is associated with the extensor digitorum longus',
           'According to the Centers for Disease Control and Prevention (2010), approximately 30 percent of all injury-related deaths in the United States are caused by head injuries',
           'The sciatic nerve branches typically split from each other in the superior portion of the popliteal fossa',
           'During running, the force applied to each foot as it contacts the ground can be up to 2.5 times your body weight']

question_prompt = """\
You are a teacher preparing a test. Please create a question that can be answered by referencing the following context.

Context:
{context}
"""

question_prompt = ChatPromptTemplate.from_template(question_prompt)

ground_truth_prompt = """\
Use the following context and question to answer this question using *only* the provided context.

Question:
{question}

Context:
{context}
"""

ground_truth_prompt = ChatPromptTemplate.from_template(ground_truth_prompt)

question_chain = question_prompt | chat_model | StrOutputParser()
ground_truth_chain = ground_truth_prompt | chat_model | StrOutputParser()
print(contexts)

i = 0
for i in range(5):
  questions.append(question_chain.invoke({"context" : documents[0].page_content}))
  contexts.append([documents[0].page_content])
  ground_truths_semantic.append(ground_truth_chain.invoke({"question" : questions[i], "context" : contexts[i]}))
  answers.append(semantic_rag_chain.invoke(questions[i]))
  i += 1
```

```python
from datasets import load_dataset, Dataset

qagc_list = []

for question, answer, context, ground_truth in zip(questions, answers, contexts, ground_truths_semantic):
  qagc_list.append({
      "question" : question,
      "answer" : answer,
      "contexts" : context,
      "ground_truth" : ground_truth
  })

eval_dataset = Dataset.from_list(qagc_list)
eval_dataset
```

```python
from ragas.metrics import (
    answer_relevancy,
    faithfulness,
    context_recall,
    context_precision,
)

#
from ragas import evaluate
import os
import openai
os.environ['OPENAI_API_KEY'] = openai_api_key
openai.api_key = os.environ['OPENAI_API_KEY']
result = evaluate(
    eval_dataset,
    metrics=[
        context_precision,
        faithfulness,
        answer_relevancy,
        context_recall,
    ],
    llm=chat_model,
    embeddings=embed_model,
    raise_exceptions=False
)
```

```python
print(result)
results_df = result.to_pandas()
results_df
```

```python
import tqdm
questions = []
ground_truths_semantic = []
contexts = []
answers = []
i = 0
for chunk in tqdm.tqdm(synthetic_data_chunks):
  questions.append(question_chain.invoke({"context" : chunk.page_content}))
  contexts.append([chunk.page_content])
  ground_truths_semantic.append(ground_truth_chain.invoke({"question" : questions[i], "context" : contexts[i]}))
  answers.append(naive_rag_chain.invoke(questions[i]))
  i += 1
```

```python
qagc_list = []

for question, answer, context, ground_truth in zip(questions, answers, contexts, ground_truths_semantic):
  qagc_list.append({
      "question" : question,
      "answer" : answer,
      "contexts" : context,
      "ground_truth" : ground_truth
  })

naive_eval_dataset = Dataset.from_list(qagc_list)
naive_eval_dataset
```

```python
from ragas.metrics import (
    answer_relevancy,
    faithfulness,
    context_recall,
    context_precision,
)

#
from ragas import evaluate
naive_result = evaluate(
    naive_eval_dataset,
    metrics=[
        context_precision,
        faithfulness,
        answer_relevancy,
        context_recall,
    ],
    llm=chat_model,
    embeddings=embed_model,
    raise_exceptions=False
)
```

```python
print(naive_result)
naive_results_df = naive_result.to_pandas()
naive_results_df
```
